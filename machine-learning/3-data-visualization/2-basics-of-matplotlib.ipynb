## 3D projections

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
```

```python
fig = plt.figure()
ax = Axes3D(fig)
x = np.arange(-5, 5, 0.25)
y = np.arange(-5, 5, 0.25)
x, y = np.meshgrid(x, y)
r = np.sqrt(x ** 2 + y ** 2)
z = np.sin(r)
```

```python
surf = ax.plot_surface(x, y, z, rstride = 1, cstride = 1, cmap = cm.coolwarm)
```

```python
plt.show()
```
