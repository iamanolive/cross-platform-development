## dataframes

```python
df = pd.DataFrame({
    "a" : [1, 2, 3],
    "b" : [4, 5, 6],
    "c" : [7, 8, 9]},
    index = [1, 2, 3])
```

```python
df = pd.DataFrame([[1, 4, 7], [2, 5, 8], [3, 6, 9]],
                  index = [1, 2, 3],
                  columns = ["a", "b", "c"])
```

```python
df.loc[3, :]
```

```python
df = pd.DataFrame([ [1, 4, 4], [2, 5, 8], [5, 6, 1]],
                  index = [1, 2, 3],
                  columns = ["a", "b", "c"])
```

```python
df
```

## duplicate values

```python
df.drop_duplicates()
```

```python
df = pd.DataFrame(
    [[1, 4, 4], [1, 5, 8], [5, 6, 1]],
    index = [1, 2, 3],
    columns = ["a", "b", "c"])
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(subset = "a")
```

```python
df
```
